```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import Pitch,VerticalPitch
import matplotlib
```

```python
text_color='w'
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mckayjohns/youtube-videos/main/data/shotmaps.csv')
```

```python
df
```

```python
dfsaved= df[df['outcome']=='Saved']
dfmissed = df[df['outcome']=='Missed']
dfpost = df[df['outcome']=='Post']
```

```python
pitch = VerticalPitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc',half=True)
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=False, tight_layout=True)
fig.set_facecolor('#22312b')

pitch.draw(ax=ax)

plt.gca().invert_yaxis()

plt.scatter(dfsaved['x'],dfsaved['y'],c='orange',s= 100,alpha=0.6,label='Saved')
plt.scatter(dfmissed['x'],dfmissed['y'],c='green',s= 100,alpha=0.6,label='Missed')
plt.scatter(dfpost['x'],dfpost['y'],c='red',s= 100,alpha=0.6,label='Post')

plt.legend(loc='upper right')

plt.title('Barcelona shots vs Juvetus',fontsize=24, c=text_color)

total_shots=len(df)

fig.text(s=f'Total Shots: {total_shots}',
        x=.27, y =.67, fontsize=20,fontfamily='Andale Mono',color=text_color)
```
