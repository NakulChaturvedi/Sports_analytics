```python
import pandas as pd
from mplsoccer.pitch import Pitch
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pitch= Pitch(pitch_color = 'grass', line_color='white', stripe=True)
fig,ax= pitch.draw()
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/mckayjohns/youtube-videos/main/data/messibetis.csv")
```

```python
df
```

```python
df['x']= df['x']*1.2
df['y']= df['y']*.8
df['endX']= df['endX']*1.2
df['endY']= df['endY']*.8
```

```python
df
```

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=False, tight_layout=True)
fig.set_facecolor('#22312b')

pitch.draw(ax=ax)
plt.gca().invert_yaxis()
pitch.kdeplot( df['x'],df['y'],ax=ax,shade=True,shade_lowest=False,alpha=0.3,n_levels=10,cmap='magma')

plt.xlim(0,120)
plt.ylim(0,80)
for x in range(len(df['x'])):
    if df ['outcome'][x] == 'Successful':
        plt.plot((df['x'][x],df['endX'][x]),(df['y'][x],df['endY'][x]), color='orange')
        plt.scatter((df['x'][x]),(df['y'][x]),color='orange',label='Successful')
for x in range(len(df['x'])):
    if df ['outcome'][x] == 'Unsuccessful':
        plt.plot((df['x'][x],df['endX'][x]),(df['y'][x],df['endY'][x]), color='red')
        plt.scatter((df['x'][x]),(df['y'][x]),color='red',label='Unsuccessful')
ax_title= ax.set_title('Messi vs Betis passes',fontsize=30,color='white')
plt.legend(loc='lower left')

```
