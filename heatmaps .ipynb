```python
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
import mplsoccer
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Chaitanya98/Football-Analytics/main/Mad%20About%20Sports/Data/Real%20Madrid%20vs%20FC%20Barcelona%202015-16.csv')
```

```python
df.head()
```

```python
dfReal= df[(df['team']=='Real Madrid') & (df['outcomeType']== 'Successful')]
```

```python
customcmap= matplotlib.colors.LinearSegmentedColormap.from_list('custom cmap',['black','green'])
```

```python
pitch=mplsoccer.Pitch(pitch_color='Black',line_color='White',line_zorder=2)
fig,ax=pitch.draw(figsize=(16,11), constrained_layout=False, tight_layout=True)
fig.set_facecolor('#22312b')
pitch.draw(ax=ax)

pitch_title=ax.set_title('Real Madrid Successful events vs Barca',fontsize=40,color='white')

pitch.kdeplot(dfReal['x'],dfReal['y'],ax=ax,shade=True,zorder=1,cmap=customcmap,n_levels=1000)
```
